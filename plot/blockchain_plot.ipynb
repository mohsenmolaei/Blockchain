```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
```

```python
date = np.array([])
time = np.array([], dtype=int)
bits = np.array([], dtype=int)
n_tx = np.array([], dtype=int)
size = np.array([], dtype=int)
fee  = np.array([], dtype=int)
weight  = np.array([], dtype=int)
height = np.array([], dtype=int)
```

```python
data= pd.read_json("https://blockchain.info/latestblock")
for counter in range(99,-1,-1):
    blk= pd.read_json("https://blockchain.info/block-height/"+str(data['height'][0] - counter)+"?format=json")

    date = np.append(date, datetime.datetime.fromtimestamp(blk["blocks"][0]['time']+16200))
    time= np.append(time,blk["blocks"][0]['time'])
    bits= np.append(bits,blk["blocks"][0]['bits'])
    n_tx= np.append(n_tx,blk["blocks"][0]['n_tx'])
    size= np.append(size,blk["blocks"][0]['size'])
    fee = np.append(fee,blk["blocks"][0]['fee'])
    weight = np.append(weight,blk["blocks"][0]['weight'])
    height= np.append(height,blk["blocks"][0]['height'])
    
    print((100-counter),'%') if (counter%10)==0  else 0
 

block_num= data['height'][0] 
```

```python
plt.figure(figsize=(15,7))
plt.plot( range(block_num-99,block_num+1), bits, color='g', label='Bits')
plt.xlabel('Number of Block', fontsize=14)
plt.ylabel('Bits (difficulty target)', fontsize=14)
plt.xticks(range(block_num-99,block_num+1,10))
plt.title("The “bits” is the difficulty target which has been encoded into a numeric value (last 100 blocks)")
plt.legend()
plt.grid()

plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.plot( range(block_num-99,block_num), np.diff(time)/60, color='g', label='Time to creat the Block')
plt.xlabel('Number of Block', fontsize=14)
plt.ylabel('Time to creat the Block (Minutes)', fontsize=14)
plt.xticks(range(block_num-99,block_num+1,10))
plt.title("Time to creat the Block (last 100 blocks in Bitcoin blockchain)")
plt.legend()
plt.grid()

plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.plot( range(block_num-99,block_num+1), size/(1024*1024), color='g', label='size')
plt.xlabel('Number of Block (current block : '+str(block_num)+')', fontsize=14)
plt.ylabel('Size (MB)', fontsize=14)
plt.xticks(range(block_num-99,block_num+1,10))
plt.title("Size of last 100 blocks in Bitcoin blockchain")
plt.legend()
plt.grid()

plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.plot( range(block_num-99,block_num+1), n_tx, color='g', label='number of transactions')
plt.xlabel('Number of Block', fontsize=14)
plt.ylabel('number of transactions', fontsize=14)
plt.xticks(range(block_num-99,block_num+1,10))
plt.title("Total Transactions in Block  (last 100 blocks in Bitcoin blockchain)")
plt.legend()
plt.grid()

plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.plot( range(block_num-99,block_num+1), fee, color='g', label='Fee')
plt.xlabel('Number of Block', fontsize=14)
plt.ylabel('Total transaction fee', fontsize=14)
plt.xticks(range(block_num-99,block_num+1,10))
plt.title("Total transaction fee for each Block  (last 100 blocks in Bitcoin blockchain)")
plt.legend()
plt.grid()

plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.plot( range(block_num-99,block_num+1), weight, color='g', label='Weight')
plt.xlabel('Total block weight', fontsize=14)
plt.ylabel('Total block weight', fontsize=14)
plt.xticks(range(block_num-99,block_num+1,10))
plt.title("Total block weight  (last 100 blocks in Bitcoin blockchain)")
plt.legend()
plt.grid()

plt.show()
```
